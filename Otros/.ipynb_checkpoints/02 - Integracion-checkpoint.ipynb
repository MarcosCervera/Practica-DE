```python
# Base SC
sc_pen["UN"] = "SC"
sc_pen['PEN'] = sc_pen['PEN'].fillna(0).astype(np.int64)
sc_pen['PEN_Org'] = sc_pen['PEN_Org'].fillna(0).astype(np.int64)

# Corrije nombre cuanto cuit es null
sc_pen['PasNombre'] = np.where(  sc_pen['PasCuit'].isnull() , 'San Cristobal' , sc_pen['PasNombre'] )
sc_pen['PasCuit'] = np.where(  sc_pen['PasCuit'].isnull() , '34500045339' , sc_pen['PasCuit'] )
```

# Base SupraPAS SC
A partir de la base sc_pen crea sc_supra. Es una base catalogo de PAS-SupraPAS (San Cristóbal). Union all de PAS y Organizadores. Si un cuit tiene 2 SupraPAS se queda con el que tiene mas pen

```python
# Base SupraPAS SC
sc_supra = sc_pen[~sc_pen["PasCuit"].isna()]
sc_supra["SupraCuit"] = np.where(sc_supra["SupraCuit"].isnull() , sc_supra["PasCuit"]  , sc_supra["SupraCuit"] )
sc_supra["SupraNombre"] = np.where(sc_supra["SupraNombre"].isnull() , sc_supra["PasNombre"]  , sc_supra["SupraNombre"] )
sc_supra = sc_supra.groupby(["PasCuit", "PasNombre", "SupraCuit", "SupraNombre"]
                              , as_index=False, dropna=False).agg({'PEN':'sum'}).sort_values(['PEN'], ascending=[ False])
sc_supra = sc_supra.drop_duplicates(['PasCuit'], keep='first')
sc_supra = sc_supra.drop(['PEN'], axis=1)

# esto se hace por los organizadores que no son SupraPAS
sc_supra_org = sc_pen[["OrgSupraCuit" , "OrgSupraNombre"]]
sc_supra_org = sc_supra_org[~sc_supra_org["OrgSupraCuit"].isna()]
sc_supra_org = sc_supra_org.drop_duplicates(['OrgSupraCuit'], keep='first')
sc_supra_org["PasCuit"] = sc_supra_org["OrgSupraCuit"]
sc_supra_org["PasNombre"] = sc_supra_org["OrgSupraNombre"]
sc_supra_org = sc_supra_org.rename(columns={'OrgSupraCuit':'SupraCuit' , 'OrgSupraNombre':'SupraNombre' })

# Junta Supra + Org
sc_supra = pd.concat([sc_supra,sc_supra_org])
sc_supra = sc_supra.drop_duplicates(['PasCuit'], keep='first')
```

# Base SupraPAS Aso
A partir de la base aso_pen crea aso_supra. Base catalogo de PAS-SupraPAS (Asociart). Trae el SupraPAS y SupraOrg desde sc_supra. Crea canal tradicional, PEN_Org Union all de PAS y Organizadores.

```python
aso_pen['PEN'] = aso_pen['PEN'].fillna(0).astype(np.int64)

# Le pone San Cristobal a Sin Organizador
aso_pen['OrgCuit'] = np.where(  aso_pen['OrgCuit'].isnull() , '34500045339' , aso_pen['OrgCuit'] )
aso_pen['OrgNombre'] = np.where(  aso_pen['OrgNombre'].isnull() , 'San Cristobal' , aso_pen['OrgNombre'] )

# Une SupraPAS
aso_pen = pd.merge(aso_pen, sc_supra[['PasCuit','SupraCuit','SupraNombre']] , on='PasCuit',  how='left')
# Une SupraOrg
aso_pen = pd.merge(aso_pen, sc_supra[['PasCuit','SupraCuit','SupraNombre']] , left_on='OrgCuit', right_on='PasCuit',  how='left')

# Trabaja campos
aso_pen["SupraCuit"] = np.where(aso_pen["SupraCuit_x"].isna() , aso_pen["PasCuit_x"] , aso_pen["SupraCuit_x"])
aso_pen["SupraNombre"] = np.where(aso_pen["SupraCuit_x"].isna() , aso_pen["PasNombre"] , aso_pen["SupraNombre_x"])
aso_pen["OrgSupraCuit"] = np.where(aso_pen["SupraCuit_y"].isna() , aso_pen["OrgCuit"] , aso_pen["SupraCuit_y"])
aso_pen["OrgSupraNombre"] = np.where(aso_pen["SupraCuit_y"].isna() , aso_pen["OrgNombre"] , aso_pen["SupraNombre_y"])

aso_pen["Flg_org_3ros"] = np.where( ( aso_pen["OrgCuit"].isna() )  | 
                               (aso_pen["OrgNombre"] == "SEGUROS DIRECTOS") | 
                               (aso_pen["OrgSupraCuit"] == aso_pen["SupraCuit"])  , 0 , 1 )
aso_pen["PEN_Org"] = aso_pen["Flg_org_3ros"] * aso_pen["PEN"]

aso_pen["Canal_N1"] = np.where( ( aso_pen["Canal"] == "Productor" )  | 
                                (aso_pen["Canal"] == "Asesor Comercial") | 
                                (aso_pen["Canal"] == "Productor en Socia") | 
                                (aso_pen["Canal"] == "Contadores" ) | 
                                (aso_pen["Canal"] == "Asociart Servicios" ) , "Canales Tradicionales" , "Otros Asociart" )
aso_pen["Canal_N2"] = np.where( ( aso_pen["Canal"] == "Productor" )  | 
                                (aso_pen["Canal"] == "Asesor Comercial") | 
                                (aso_pen["Canal"] == "Productor en Socia") | 
                                (aso_pen["Canal"] == "Contadores" )  , "Tradicional" , 
                      np.where( (aso_pen["Canal"] == "Asociart Servicios" ) , "Directo" , 
                      np.where( (aso_pen["Canal"] == "Agente Institorio" ) , "Bancos" , "Otros Asociart" ) ) )

aso_pen["UN"] = "Aso"

aso_pen = aso_pen.rename(columns={'PasCuit_x':'PasCuit'})
aso_pen = aso_pen.drop(['SupraCuit_x', 'SupraNombre_x', 'SupraCuit_y', 'SupraNombre_y','PasCuit_y'], axis=1)


# Arma SupraPAS Aso (para agregar PAS y SupraPAS que trabajan solo en Aso al listado final)
aso_supra = aso_pen[["PasCuit" , "PasNombre" , "SupraCuit" , "SupraNombre"]]
aso_supra = aso_supra[~aso_supra["PasCuit"].isna()]
aso_supra = aso_supra.drop_duplicates(['PasCuit'], keep='first')

# esto se hace por los organizadores que no son SupraPAS
aso_supra_org = aso_pen[aso_pen["Canal_N2"] == "Tradicional"]
aso_supra_org = aso_supra_org[["OrgSupraCuit" , "OrgSupraNombre"]]
aso_supra_org = aso_supra_org[~aso_supra_org["OrgSupraCuit"].isna()]
aso_supra_org = aso_supra_org.drop_duplicates(['OrgSupraCuit'], keep='first')
aso_supra_org["PasCuit"] = aso_supra_org["OrgSupraCuit"]
aso_supra_org["PasNombre"] = aso_supra_org["OrgSupraNombre"]
aso_supra_org = aso_supra_org.rename(columns={'OrgSupraCuit':'SupraCuit' , 'OrgSupraNombre':'SupraNombre' })

# Junta Supra + Org
aso_supra = pd.concat([aso_supra,aso_supra_org])
aso_supra = aso_supra.drop_duplicates(['PasCuit'], keep='first')
```

# Base SupraPAS Retiro

```python
# Une SupraPAS
ret_pen = pd.merge(retiro_pas, sc_supra[['PasCuit','SupraCuit','SupraNombre']] , on='PasCuit',  how='left')
ret_pen["SupraCuit"] = np.where(ret_pen["SupraCuit"].isna() , ret_pen["PasCuit"] , ret_pen["SupraCuit"])
ret_pen["SupraNombre"] = np.where(ret_pen["SupraNombre"].isna() , ret_pen["PasNombre"] , ret_pen["SupraNombre"])

ret_pen["UN"] = "Ret"

# Arma SupraPAS Ret (para agregar PAS y SupraPAS que trabajan solo en Aso al listado final)
ret_supra = ret_pen[["PasCuit" , "PasNombre" , "SupraCuit" , "SupraNombre"]]
ret_supra = ret_supra[~ret_supra["PasCuit"].isna()]
ret_supra = ret_supra.drop_duplicates(['PasCuit'], keep='first')
ret_supra["SupraNombre"] = ret_supra["SupraNombre"].str.upper()
ret_supra["PasNombre"] = ret_supra["PasNombre"].str.upper()
```

# Base SupraPAS GSC
Es una base catalogo PAS_SupraPAS

```python
pas_gsc_supra = pd.concat([sc_supra,aso_supra,ret_supra])
pas_gsc_supra = pas_gsc_supra.drop_duplicates(['PasCuit'], keep='first')
pas_gsc_supra = pas_gsc_supra.replace(',', '' , regex=True )
```

```python
pas_gsc_supra_segm = pd.merge(pas_gsc_supra, segmentacion_pas[['PasCuit','Segmento','Segmento_ord']]  , left_on='SupraCuit', right_on='PasCuit',  how='left')
pas_gsc_supra_segm = pas_gsc_supra_segm.drop(['PasCuit_y'], axis=1)
pas_gsc_supra_segm = pas_gsc_supra_segm.rename(columns={'PasCuit_x':'PasCuit'})
pas_gsc_supra_segm = pas_gsc_supra_segm.drop_duplicates(['PasCuit'], keep='first')
```

- Agraria

Agrega a la base SupraPAS GSC el segmento y si el pas está vinculado con Agraria

```python
sc_agraria = sc_pen[sc_pen["Canal"] == "AGRARIA"] 
sc_agraria = sc_agraria[["PasCuit"]]
sc_agraria["Flg_Agraria"] = 1
sc_agraria = sc_agraria.drop_duplicates(['PasCuit'], keep='first')

sc_agraria_org = sc_pen[sc_pen["Canal"] == "AGRARIA"] 
sc_agraria_org = sc_agraria_org[["OrgSupraCuit"]]
sc_agraria_org["Flg_Agraria"] = 1
sc_agraria_org = sc_agraria_org.drop_duplicates(['OrgSupraCuit'], keep='first')
sc_agraria_org = sc_agraria_org.rename(columns={'OrgSupraCuit':'PasCuit' })

sc_agraria = pd.concat([sc_agraria,sc_agraria_org])
sc_agraria = sc_agraria.drop_duplicates(['PasCuit'], keep='first')

# Junta con segmento
pas_gsc_supra_segm = pd.merge(pas_gsc_supra_segm, sc_agraria  , on='PasCuit',  how='left')
pas_gsc_supra_segm['Flg_Agraria'] = pas_gsc_supra_segm['Flg_Agraria'].fillna(0).astype(np.int64)
```

- Otros canales

Agrega a la base SupraPAS GSC el indicador que nos son canal tradicional y grandes brokers
Para los casos que tienen 2 canales en sc, toma el que tiene mas PEN

```python
# San Cristobal
sc_pas_canales = sc_pen[~sc_pen["PasCuit"].isna()]
sc_pas_canales = sc_pas_canales.groupby(["PasCuit", "PasNombre", "SupraCuit", "SupraNombre", "Canal", "Canal_N2"]
                              , as_index=False, dropna=False).agg({'PEN':'sum'}).sort_values(['PEN'], ascending=[ False])
sc_pas_canales = sc_pas_canales.drop_duplicates(['PasCuit'], keep='first')
sc_pas_canales = sc_pas_canales.drop(['PEN'], axis=1)
sc_pas_canales = sc_pas_canales[sc_pas_canales["Canal_N2"] != "Tradicional"]
sc_pas_canales = sc_pas_canales[sc_pas_canales["Canal"] != "GRANDES BROKERS"]
sc_pas_canales = sc_pas_canales[~sc_pas_canales["Canal_N2"].isna()]
sc_pas_canales['Otros_canales'] = 1

# Asociart
aso_pas_canales = aso_pen[~aso_pen["PasCuit"].isna()]
aso_pas_canales = aso_pas_canales.groupby(["PasCuit", "PasNombre", "SupraCuit", "SupraNombre", "Canal_N2"]
                              , as_index=False, dropna=False).agg({'PEN':'sum'}).sort_values(['PEN'], ascending=[ False])
aso_pas_canales = aso_pas_canales.drop_duplicates(['PasCuit'], keep='first')
aso_pas_canales = aso_pas_canales.drop(['PEN'], axis=1)
aso_pas_canales = aso_pas_canales[aso_pas_canales["Canal_N2"] != "Tradicional"]
aso_pas_canales['Otros_canales'] = 1

# GSC
gsc_pas_canales = pd.concat([sc_pas_canales,aso_pas_canales])
gsc_pas_canales = gsc_pas_canales.drop_duplicates(['PasCuit'], keep='first')


# Junta con segmento
pas_gsc_supra_segm = pd.merge(pas_gsc_supra_segm, gsc_pas_canales[['PasCuit','Otros_canales']]  , on='PasCuit',  how='left')
pas_gsc_supra_segm['Otros_canales'] = pas_gsc_supra_segm['Otros_canales'].fillna(0).astype(np.int64)
```

- Base Final

```python
# Agrega segmentos a los vacios
pas_gsc_supra_segm['Segmento'] = np.where(  ~pas_gsc_supra_segm['Segmento'].isnull() ,pas_gsc_supra_segm['Segmento']  ,
                                  np.where(  pas_gsc_supra_segm["Flg_Agraria"] == 1 , "Agraria" ,
                                  np.where(  pas_gsc_supra_segm["Otros_canales"] == 1, "No Segmentado" , "Origen" ) ) )

pas_gsc_supra_segm['Segmento_ord'] = np.where(  pas_gsc_supra_segm['Segmento'] == 'Elite +' , '1-Elite +' ,
                                     np.where(  pas_gsc_supra_segm['Segmento'] == 'Elite' , '2-Elite' ,
                                     np.where(  pas_gsc_supra_segm['Segmento'] == 'Evolución' , '3-Evolución' ,
                                     np.where(  pas_gsc_supra_segm['Segmento'] == 'Origen' , '4-Origen' ,
                                     np.where(  pas_gsc_supra_segm['Segmento'] == 'Agraria' , '5-Agraria' , 
                                              '9-No Segmentado' )))))    

pas_gsc_supra_segm = pas_gsc_supra_segm.replace(',', '' , regex=True )

pas_gsc_segm = pas_gsc_supra_segm.drop(['SupraCuit','SupraNombre', 'Flg_Agraria', 'Otros_canales'], axis=1)
```

# Base Fecha
Base AnoMes con flag A1 y A2

```python
# creo base fecha
# aso_pen = aso_pen.rename(columns={'Periodo':'AnoMes'})
d_fecha = aso_pen.groupby(["AnoMes"], as_index=False).size().sort_values(['AnoMes'], ascending=[ False]).reset_index(level=0, inplace=False) 
d_fecha=d_fecha.drop(["index"], axis=1)
d_fecha["Indice"] = d_fecha.index + 1
d_fecha["A1_Flg"] = np.where(d_fecha["Indice"] < 13 , 1 , 0) 
d_fecha["A2_Flg"] = np.where( (d_fecha["Indice"] > 12) & (d_fecha["Indice"] < 25) , 1 , 0) 
```

# Base GSC

```python
# Normaliza nombre de PAS
aso_pen = aso_pen.drop(['PasNombre'], axis=1)
aso_pen = pd.merge(aso_pen, pas_gsc_supra_segm[['PasCuit','PasNombre']]  , on='PasCuit',  how='left')

# Tablas intermedias de SC y Aso para armar base GSC
aso_pen_int = aso_pen.groupby(["AnoMes", "PasCod",  "PasCuit", "PasNombre", "Canal", "Canal_N1", "Canal_N2","SupraCuit", 
                               "SupraNombre", "OrgSupraCuit", "OrgSupraNombre", "LinkerCuit", "LinkerNombre","Flg_org_3ros",
                               "OrgCuit",  "OrgNombre", "UN", "EjecutivoCuenta", "Sucursal", "Region", "SubRegion", "Ramo" ,
                               "SubRamo"], as_index=False, dropna=False).agg({'PEN':'sum','PEN_Org':'sum'})


sc_pen_int = sc_pen[['AnoMes', 'Canal', 'Canal_N2', 'Canal_N1', 'Sucursal', 'Ramo',
       'SubRamo', 'PasCod', 'PasCuit', 'PasNombre', 
       'SupraCuit', 'SupraNombre', 'OrgCuit', 'OrgNombre', 'OrgSupraCuit',
       'OrgSupraNombre', 'LinkerCuit', 'LinkerNombre', 'Flg_Agraria',
       'Flg_org_3ros', 'EjecutivoCuenta', 'PEN', 'PEN_Org', 'UN']]

gsc_pen = pd.concat([sc_pen_int,aso_pen_int])

# Cambia nombre de Sin Organizador
gsc_pen['OrgSupraNombre'] = np.where(  gsc_pen['OrgSupraNombre'] == 'San Cristobal' , 'Sin Organizador' ,  gsc_pen['OrgSupraNombre'] )
gsc_pen['OrgSupraCuit'] = np.where(  gsc_pen['OrgSupraCuit'] == '34500045339' , '99999999999' ,  gsc_pen['OrgSupraCuit'] )


# indicadores de PEN

gsc_pen = pd.merge(gsc_pen, d_fecha[['AnoMes','A1_Flg','A2_Flg']] , on='AnoMes',  how='left')

gsc_pen["PEN_GSC_Sup_A1"] = gsc_pen["PEN"] * gsc_pen["A1_Flg"] 
gsc_pen["PEN_GSC_Sup_A2"] = gsc_pen["PEN"] * gsc_pen["A2_Flg"] 
gsc_pen["PEN_GSC_Org_A1"] = gsc_pen["PEN_Org"] * gsc_pen["A1_Flg"] 
gsc_pen["PEN_GSC_Org_A2"] = gsc_pen["PEN_Org"] * gsc_pen["A2_Flg"] 
gsc_pen["PEN_SC"] = np.where(gsc_pen["UN"] == "SC" , gsc_pen["PEN"] , 0 )
gsc_pen["PEN_SC_Sup_A1"] = np.where(gsc_pen["UN"] == "SC" , gsc_pen["PEN"] * gsc_pen["A1_Flg"] , 0 )
gsc_pen["PEN_SC_Sup_A2"] = np.where(gsc_pen["UN"] == "SC" , gsc_pen["PEN"] * gsc_pen["A2_Flg"] , 0 )
gsc_pen["PEN_SC_Org_A1"] = np.where(gsc_pen["UN"] == "SC" , gsc_pen["PEN_Org"] * gsc_pen["A1_Flg"] , 0 ) 
gsc_pen["PEN_SC_Org_A2"] = np.where(gsc_pen["UN"] == "SC" , gsc_pen["PEN_Org"] * gsc_pen["A2_Flg"] , 0 )
gsc_pen["PEN_Aso"] = np.where(gsc_pen["UN"] == "Aso" , gsc_pen["PEN"] , 0 )
gsc_pen["PEN_Aso_Sup_A1"] = np.where(gsc_pen["UN"] == "Aso" , gsc_pen["PEN"] * gsc_pen["A1_Flg"] , 0 )
gsc_pen["PEN_Aso_Sup_A2"] = np.where(gsc_pen["UN"] == "Aso" , gsc_pen["PEN"] * gsc_pen["A2_Flg"] , 0 )
gsc_pen["PEN_Aso_Org_A1"] = np.where(gsc_pen["UN"] == "Aso" , gsc_pen["PEN_Org"] * gsc_pen["A1_Flg"] , 0 )
gsc_pen["PEN_Aso_Org_A2"] = np.where(gsc_pen["UN"] == "Aso" , gsc_pen["PEN_Org"] * gsc_pen["A2_Flg"] , 0 )

gsc_pen = gsc_pen[gsc_pen['PEN'] != 0 ] 
gsc_pen = gsc_pen.replace(',', '' , regex=True )
```

```python
# Base PAS
gsc_pen_pas = gsc_pen.copy()
gsc_pen_pas["A_Cuit"] = gsc_pen_pas["PasCuit"]
gsc_pen_pas["A_Nombre"] = gsc_pen_pas["PasNombre"]
gsc_pen_pas["A_Rol"] = "PAS"
gsc_pen_pas = gsc_pen_pas.drop(['PEN_Org','PEN_GSC_Org_A1', 'PEN_GSC_Org_A2', 'PEN_SC_Org_A1', 'PEN_SC_Org_A2', 'PEN_Aso_Org_A1', 'PEN_Aso_Org_A2'], axis=1)

# Base SupraPAS
gsc_pen_supra = gsc_pen.copy()
gsc_pen_supra["A_Cuit"] = gsc_pen_supra["SupraCuit"]
gsc_pen_supra["A_Nombre"] = gsc_pen_supra["SupraNombre"]
gsc_pen_supra["A_Rol"] = "SupraPAS"
gsc_pen_supra = gsc_pen_supra.drop(['PEN_Org','PEN_GSC_Org_A1', 'PEN_GSC_Org_A2', 'PEN_SC_Org_A1', 'PEN_SC_Org_A2', 'PEN_Aso_Org_A1', 'PEN_Aso_Org_A2'], axis=1)
  
# Base Organizador    
gsc_pen_org = gsc_pen.copy()
gsc_pen_org["A_Cuit"] = gsc_pen_org["OrgSupraCuit"]
gsc_pen_org["A_Nombre"] = gsc_pen_org["OrgSupraNombre"]
gsc_pen_org["A_Rol"] = "Organizador"
gsc_pen_org = gsc_pen_org.drop(['PEN','PEN_SC','PEN_Aso','PEN_GSC_Sup_A1', 'PEN_GSC_Sup_A2', 'PEN_SC_Sup_A1', 'PEN_SC_Sup_A2', 'PEN_Aso_Sup_A1', 'PEN_Aso_Sup_A2'], axis=1)

gsc_pen_rol = pd.concat([gsc_pen_pas,gsc_pen_supra,gsc_pen_org])
gsc_pen_rol = gsc_pen_rol.replace(',', '' , regex=True )

gsc_pen_rol = gsc_pen_rol[~((gsc_pen_rol['A_Rol']== "Organizador") & (gsc_pen_rol['PEN_Org']== 0))] 
```
